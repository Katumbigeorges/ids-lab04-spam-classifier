```python
"""
Lab 04 - Part I: Dataset Generation
04-625 A: Intrusion Detection Systems
Author: ggeorges
"""

import random
import os

# ──────────────────────────────────────────────
# 1.  DICTIONARIES
# ──────────────────────────────────────────────

SPAM_DICT = [
    "free", "money", "click", "discount", "offer", "win", "prize", "cash",
    "earn", "buy", "sale", "deal", "limited", "urgent", "act", "now",
    "guarantee", "credit", "loan", "investment", "bonus", "reward",
    "congratulations", "selected", "exclusive", "risk", "profit", "income",
    "wealth", "cheap", "affordable", "subscribe", "unsubscribe", "promo",
    "giveaway", "lottery", "jackpot", "winner", "claim", "redeem", "coupon",
    "savings", "percent", "off", "viagra", "pills", "pharmacy", "casino",
    "bet", "gamble", "crypto", "bitcoin", "inheritance", "million",
]

HAM_DICT = [
    "meeting", "report", "schedule", "project", "update", "agenda",
    "review", "document", "presentation", "deadline", "budget", "team",
    "client", "proposal", "feedback", "analysis", "strategy", "planning",
    "conference", "follow", "up", "attached", "please", "find", "regards",
    "sincerely", "draft", "version", "approval", "request", "invoice",
    "contract", "agreement", "discussion", "notes", "minutes", "action",
    "items", "progress", "status", "quarterly", "annual", "weekly",
    "training", "onboarding", "interview", "candidate", "position",
    "department", "office", "colleague", "manager", "director",
]

# ──────────────────────────────────────────────
# 2.  EMAIL GENERATION
# ──────────────────────────────────────────────

PS = 0.75   # P(spam word | spam email)
QS = 0.20   # P(spam word | ham email)
L_MIN = 5
L_MAX = 15

N_SPAM = 1000
N_HAM  = 2300


def generate_email(spam_dict, ham_dict, p_spam, min_length, max_length):
    """Generate a synthetic email by randomly selecting words from dictionaries."""
    email_length = random.randint(min_length, max_length)
    email_words = []
    for _ in range(email_length):
        if random.random() < p_spam:
            email_words.append(random.choice(spam_dict))
        else:
            email_words.append(random.choice(ham_dict))
    return " ".join(email_words)


def generate_dataset(n_spam, n_ham):
    """Generate full dataset of spam and ham emails."""
    spam_emails = [
        generate_email(SPAM_DICT, HAM_DICT, PS, L_MIN, L_MAX)
        for _ in range(n_spam)
    ]
    ham_emails = [
        generate_email(SPAM_DICT, HAM_DICT, QS, L_MIN, L_MAX)
        for _ in range(n_ham)
    ]
    return spam_emails, ham_emails


def save_emails(emails, filepath):
    """Save a list of emails to a text file, one per line."""
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    with open(filepath, "w") as f:
        for email in emails:
            f.write(email + "\n")
    print(f"[+] Saved {len(emails)} emails → {filepath}")


def display_samples(spam_emails, ham_emails, n=3):
    """Display a few sample emails from each category."""
    print("\n--- SPAM SAMPLES ---")
    for i, e in enumerate(spam_emails[:n], 1):
        print(f"  [{i}] {e}")
    print("\n--- HAM SAMPLES ---")
    for i, e in enumerate(ham_emails[:n], 1):
        print(f"  [{i}] {e}")


def save_dictionaries():
    """Save spam and ham dictionaries to text files."""
    os.makedirs("data", exist_ok=True)
    with open("data/spam_dict.txt", "w") as f:
        f.write("\n".join(SPAM_DICT))
    with open("data/ham_dict.txt", "w") as f:
        f.write("\n".join(HAM_DICT))
    print("[+] Dictionaries saved → data/spam_dict.txt, data/ham_dict.txt")


# ──────────────────────────────────────────────
# 3.  MAIN
# ──────────────────────────────────────────────

if __name__ == "__main__":
    random.seed(42)

    print("=" * 50)
    print("  Lab 04 - Part I: Dataset Generation")
    print("=" * 50)

    # Save dictionaries
    save_dictionaries()

    # Generate emails
    print(f"\n[*] Generating {N_SPAM} spam and {N_HAM} ham emails...")
    spam_emails, ham_emails = generate_dataset(N_SPAM, N_HAM)

    # Save to files
    save_emails(spam_emails, "data/spam_emails.txt")
    save_emails(ham_emails,  "data/ham_emails.txt")

    # Display samples
    display_samples(spam_emails, ham_emails)

    print("\n[✓] Part I complete.")
```

```python
import math
from collections import defaultdict

# ──────────────────────────────────────────────
# LOAD EMAILS
# ──────────────────────────────────────────────

def load_emails(filepath):
    with open(filepath, "r") as f:
        return [line.strip() for line in f if line.strip()]

spam_emails = load_emails("data/spam_emails.txt")
ham_emails  = load_emails("data/ham_emails.txt")
print(f"[*] Loaded {len(spam_emails)} spam, {len(ham_emails)} ham emails")

# ──────────────────────────────────────────────
# COMPUTE PROBABILITIES
# ──────────────────────────────────────────────

n_spam  = len(spam_emails)
n_ham   = len(ham_emails)
n_total = n_spam + n_ham

# Prior probabilities
p_spam = n_spam  / n_total
p_ham  = n_ham   / n_total

# Word counts
spam_counts = defaultdict(int)
ham_counts  = defaultdict(int)

for email in spam_emails:
    for word in email.split():
        spam_counts[word] += 1

for email in ham_emails:
    for word in email.split():
        ham_counts[word] += 1

# Vocabulary
vocabulary = set(spam_counts.keys()) | set(ham_counts.keys())
V = len(vocabulary)

total_spam_words = sum(spam_counts.values())
total_ham_words  = sum(ham_counts.values())

# Laplace smoothing + likelihood ratios
word_likelihoods = {}
p_word_spam      = {}
p_word_ham       = {}

for word in vocabulary:
    p_w_spam = (spam_counts[word] + 1) / (total_spam_words + V)
    p_w_ham  = (ham_counts[word]  + 1) / (total_ham_words  + V)
    p_word_spam[word]    = p_w_spam
    p_word_ham[word]     = p_w_ham
    word_likelihoods[word] = p_w_spam / p_w_ham

print(f"[*] Prior P(spam) = {p_spam:.4f}  |  P(ham) = {p_ham:.4f}")
print(f"[*] Vocabulary size: {V} words")

# ──────────────────────────────────────────────
# CLASSIFICATION FUNCTIONS
# ──────────────────────────────────────────────

def compute_likelihood_ratio(email, word_likelihoods):
    ratio = 1.0
    for word in email.split():
        ratio *= word_likelihoods.get(word, 1.0)
    return ratio

def classify_email(email, word_likelihoods, p_spam, p_ham):
    lr = compute_likelihood_ratio(email, word_likelihoods)

    # MAP
    map_spam_score = lr * p_spam
    map_ham_score  = (1.0 / lr) * p_ham if lr != 0 else float("inf")
    map_label = "spam" if map_spam_score > map_ham_score else "ham"

    # ML
    ml_label = "spam" if lr > 1 else "ham"

    return map_label, ml_label, lr

# ──────────────────────────────────────────────
# TEST ON 10 EMAILS
# ──────────────────────────────────────────────

random.seed(99)

test_emails = []
test_labels = []
for _ in range(5):
    test_emails.append(generate_email(SPAM_DICT, HAM_DICT, PS, L_MIN, L_MAX))
    test_labels.append("spam")
for _ in range(5):
    test_emails.append(generate_email(SPAM_DICT, HAM_DICT, QS, L_MIN, L_MAX))
    test_labels.append("ham")

print("\n" + "=" * 70)
print("  Part II — Classification Results on 10 Test Emails")
print("=" * 70)
print(f"{'#':<3} {'True':<6} {'MAP':<8} {'ML':<8} {'LR':>14}  {'Email'}")
print("-" * 70)

map_correct = 0
ml_correct  = 0

for i, (email, true_label) in enumerate(zip(test_emails, test_labels), 1):
    map_pred, ml_pred, lr = classify_email(email, word_likelihoods, p_spam, p_ham)
    map_ok = "✓" if map_pred == true_label else "✗"
    ml_ok  = "✓" if ml_pred  == true_label else "✗"
    if map_pred == true_label: map_correct += 1
    if ml_pred  == true_label: ml_correct  += 1
    snippet = email[:35] + "..." if len(email) > 35 else email
    print(f"{i:<3} {true_label:<6} {map_pred+map_ok:<9} {ml_pred+ml_ok:<9} {lr:>14.4f}  {snippet}")

print("-" * 70)
print(f"MAP accuracy: {map_correct}/10  |  ML accuracy: {ml_correct}/10")

# Save test emails for Part III
with open("data/test_emails.txt", "w") as f:
    for email, label in zip(test_emails, test_labels):
        f.write(f"{label}\t{email}\n")

print("\n[✓] Test emails saved → data/test_emails.txt")
print("[✓] Part II complete.")
```

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import math
import os
os.makedirs("images", exist_ok=True)
print("[✓] Libraries ready")
```

```python
# ──────────────────────────────────────────────
# PART III - TASK 1: FEATURE EXTRACTION
# ──────────────────────────────────────────────

ham_dict_set = set(HAM_DICT)

def compute_xham(email, ham_dict_set):
    """Compute fraction of words in email that belong to ham dictionary."""
    words = email.split()
    if not words:
        return 0.0
    ham_count = sum(1 for w in words if w in ham_dict_set)
    return ham_count / len(words)

def build_feature_dataset(spam_emails, ham_emails, ham_dict_set):
    """Build (xham, label) dataset. label=1 for ham, 0 for spam."""
    X, y = [], []
    for email in spam_emails:
        X.append(compute_xham(email, ham_dict_set))
        y.append(0)
    for email in ham_emails:
        X.append(compute_xham(email, ham_dict_set))
        y.append(1)
    return np.array(X), np.array(y)

# Build the dataset
X, y = build_feature_dataset(spam_emails, ham_emails, ham_dict_set)

print(f"[*] Feature extraction complete")
print(f"    Dataset shape : {X.shape}")
print(f"    Spam emails   : {int((y==0).sum())}")
print(f"    Ham  emails   : {int((y==1).sum())}")
print(f"\n    Sample xham values (first 5 spam): {X[:5].round(3)}")
print(f"    Sample xham values (first 5 ham) : {X[1000:1005].round(3)}")
```

```python
# ──────────────────────────────────────────────
# PART III - TASK 2: SINGLE NEURON CLASSIFIER
# ──────────────────────────────────────────────

def sigmoid(z):
    """Sigmoid activation function."""
    return 1.0 / (1.0 + np.exp(-np.clip(z, -500, 500)))

def train_neuron(X, y, lr=0.5, epochs=500, seed=0):
    """Train single neuron using batch gradient descent."""
    np.random.seed(seed)
    w = np.random.randn() * 0.01
    b = np.random.randn() * 0.01

    N = len(X)
    loss_history = []

    for epoch in range(epochs):
        # Forward pass
        z     = w * X + b
        y_hat = sigmoid(z)

        # MSE Loss
        loss = np.mean((y_hat - y) ** 2)
        loss_history.append(loss)

        # Gradients
        error   = y_hat - y
        d_sigma = y_hat * (1 - y_hat)
        dL_dw   = np.mean(2 * error * d_sigma * X)
        dL_db   = np.mean(2 * error * d_sigma)

        # Update weights
        w -= lr * dL_dw
        b -= lr * dL_db

        if (epoch + 1) % 100 == 0:
            print(f"  Epoch {epoch+1:>4d} | Loss: {loss:.6f} | w: {w:.4f} | b: {b:.4f}")

    return w, b, loss_history

def compute_threshold(w, b):
    """Compute classification threshold xT."""
    y0  = sigmoid(b)
    y1  = sigmoid(w + b)
    mid = (y0 + y1) / 2.0
    mid = np.clip(mid, 1e-7, 1 - 1e-7)
    logit_mid = math.log(mid / (1 - mid))
    xT = (logit_mid - b) / w if w != 0 else 0.5
    return float(xT), float(y0), float(y1), float(mid)

# Train
print("[*] Training single neuron (batch gradient descent)...")
print("-" * 55)
w, b, loss_history = train_neuron(X, y, lr=0.5, epochs=500)

# Threshold
xT, y0, y1, mid = compute_threshold(w, b)

print("-" * 55)
print(f"\n[+] Final weights  →  w = {w:.4f},  b = {b:.4f}")
print(f"[+] Threshold xT   →  {xT:.4f}")
print(f"    y(0) = {y0:.4f}  |  y(1) = {y1:.4f}  |  midpoint = {mid:.4f}")
```

```python
# ──────────────────────────────────────────────
# PART III - TASK 3: TEST AND COMPARE
# ──────────────────────────────────────────────

def classify_neuron(xham, w, b, xT):
    """Classify using trained neuron."""
    y_out = sigmoid(w * xham + b)
    return "ham" if xham >= xT else "spam", float(y_out)

# Load test emails saved in Part II
with open("data/test_emails.txt") as f:
    lines = [l.strip().split("\t") for l in f if l.strip()]

test_labels_p3 = [l[0] for l in lines]
test_emails_p3 = [l[1] for l in lines]

print("=" * 80)
print("  Part III — Test Results and Classifier Comparison")
print("=" * 80)
print(f"{'#':<3} {'True':<6} {'NN':<8} {'MAP':<8} {'ML':<8} {'xham':>6}  {'Email'}")
print("-" * 80)

nn_preds  = []
map_preds = []
ml_preds  = []

for i, (email, true_label) in enumerate(zip(test_emails_p3, test_labels_p3), 1):
    xham = compute_xham(email, ham_dict_set)
    nn_pred,  _  = classify_neuron(xham, w, b, xT)
    map_pred, ml_pred, _ = classify_email(
        email, word_likelihoods, p_spam, p_ham
    )

    nn_preds.append(nn_pred)
    map_preds.append(map_pred)
    ml_preds.append(ml_pred)

    def mark(pred):
        return pred + ("✓" if pred == true_label else "✗")

    snippet = email[:30] + "..." if len(email) > 30 else email
    print(f"{i:<3} {true_label:<6} {mark(nn_pred):<9} {mark(map_pred):<9} "
          f"{mark(ml_pred):<9} {xham:>6.3f}  {snippet}")

print("-" * 80)
def acc(preds):
    return sum(p == t for p, t in zip(preds, test_labels_p3))

print(f"Accuracy  →  NN: {acc(nn_preds)}/10  |  "
      f"MAP: {acc(map_preds)}/10  |  ML: {acc(ml_preds)}/10")
```

```python
# Comparison bar chart
def accuracy_score(preds):
    return sum(p == t for p, t in zip(preds, test_labels_p3)) / len(test_labels_p3)

names  = ["Neural Net", "MAP Bayes", "ML Bayes"]
values = [accuracy_score(nn_preds),
          accuracy_score(map_preds),
          accuracy_score(ml_preds)]
colors = ["#2E75B6", "#ED7D31", "#70AD47"]

fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(names, values, color=colors, width=0.5)
for bar, val in zip(bars, values):
    ax.text(bar.get_x() + bar.get_width() / 2, val + 0.02,
            f"{val:.0%}", ha="center", va="bottom", fontsize=11)
ax.set_ylim(0, 1.15)
ax.set_ylabel("Accuracy")
ax.set_title("Classifier Comparison — 10 Test Emails")
ax.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.savefig("images/lab4_comparison.png", dpi=150)
plt.show()
print("[+] Saved images/lab4_comparison.png")
```
